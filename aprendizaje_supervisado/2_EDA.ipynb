```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data/titanic_clean.csv')
```

```python
plt.figure(figsize=(12, 8))

sns.barplot(x = df['Survived'].value_counts().index, y = df['Survived'].value_counts())
plt.title('Value Counts of Survived')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Eliminar valores nulos en la columna Age
edades = df["Age"].dropna()

# Crear la figura
plt.figure(figsize=(12,6))

# Histograma con curva de densidad
sns.histplot(
    edades,
    bins=30,        # número de intervalos
    kde=True        # agrega curva de densidad
)

# Personalización
plt.title("Distribución de edades de los pasajeros")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Crear la figura
plt.figure(figsize=(12,6))

# Gráfica de conteo de sobrevivientes por clase
sns.countplot(
    data=df,
    x="Pclass",
    hue="Survived"
)

# Personalización
plt.title("Cantidad de sobrevivientes por clase de pasajero")
plt.xlabel("Clase del pasajero (Pclass)")
plt.ylabel("Cantidad de pasajeros")
plt.legend(title="Sobrevivió", labels=["No", "Sí"])

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Eliminar posibles valores nulos en Embarked
df_embarked = df.dropna(subset=["Embarked"])

# Crear la figura
plt.figure(figsize=(12,6))

# Gráfica de conteo de sobrevivientes por puerto
sns.countplot(
    data=df_embarked,
    x="Embarked",
    hue="Survived"
)

# Personalización
plt.title("Cantidad de sobrevivientes por puerto de embarque")
plt.xlabel("Puerto de embarque (Embarked)")
plt.ylabel("Cantidad de pasajeros")
plt.legend(title="Sobrevivió", labels=["No", "Sí"])

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Eliminar valores nulos en Age
df_edades = df.dropna(subset=["Age"])

# Crear categorías de edad
bins = [0, 2, 12, 18, 30, 60, 100]
labels = ["Infante", "Niño", "Adolescente", "Joven Adulto", "Adulto", "Mayor"]

df_edades["Grupo_Edad"] = pd.cut(
    df_edades["Age"],
    bins=bins,
    labels=labels,
    right=False
)

# Calcular tasa de supervivencia por grupo
tasa_supervivencia = df_edades.groupby("Grupo_Edad")["Survived"].mean()

# Crear figura
plt.figure(figsize=(12,6))

# Gráfica de barras
sns.barplot(
    x=tasa_supervivencia.index,
    y=tasa_supervivencia.values
)

# Personalización
plt.title("Tasa de supervivencia por grupo de edad")
plt.xlabel("Grupo de Edad")
plt.ylabel("Tasa de Supervivencia")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Obtener edad máxima y mínima (ignorando valores nulos)
edad_maxima = df["Age"].max()
edad_minima = df["Age"].min()

# Mostrar resultados
print("Edad máxima:", edad_maxima)
print("Edad mínima:", edad_minima)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Eliminar valores nulos en Age
edades = df["Age"].dropna()

# Calcular medidas estadísticas
media = edades.mean()
mediana = edades.median()
moda = edades.mode()[0]  # puede haber varias modas, tomamos la primera

# Crear figura
plt.figure(figsize=(12,6))

# Histograma
sns.histplot(edades, bins=30, kde=True)

# Líneas verticales
plt.axvline(media, linestyle="--", label=f"Media: {media:.2f}")
plt.axvline(mediana, linestyle="-.", label=f"Mediana: {mediana:.2f}")
plt.axvline(moda, linestyle=":", label=f"Moda: {moda:.2f}")

# Personalización
plt.title("Distribución de edades con media, mediana y moda")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Calcular tasa de supervivencia por número de SibSp
tasa_supervivencia = df.groupby("SibSp")["Survived"].mean()

# Crear figura
plt.figure(figsize=(12,6))

# Gráfica de barras
sns.barplot(
    x=tasa_supervivencia.index,
    y=tasa_supervivencia.values
)

# Personalización
plt.title("Tasa de supervivencia según número de hermanos/esposos a bordo (SibSp)")
plt.xlabel("Número de hermanos/esposos a bordo (SibSp)")
plt.ylabel("Tasa de supervivencia")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Seleccionar solo columnas numéricas
df_numerico = df.select_dtypes(include=["number"])

# Calcular matriz de correlación
correlacion = df_numerico.corr()

# Crear figura
plt.figure(figsize=(12,8))

# Crear heatmap
sns.heatmap(
    correlacion,
    annot=True,        # muestra valores numéricos
    fmt=".2f",         # formato a 2 decimales
    cmap="coolwarm",   # paleta de colores
    linewidths=0.5
)

# Personalización
plt.title("Matriz de correlación de variables numéricas")

plt.tight_layout()
plt.show()
```
